```python
# Calculate the mean rating for all movies
C = ratings['rating'].mean()
C
```

```python
# Calculate the minimum number of votes required to be listed in the chart (90th percentile)
m = ratings['movieId'].value_counts().quantile(0.90)
m
```

```python
# Filter movies that qualify for the chart
qualified_movies = ratings.groupby('movieId').filter(lambda x: len(x) >= m)
qualified_movies
```

```python
# Define the weighted rating formula
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['rating']
    return (v/(v+m) * R) + (m/(m+v) * C)

# Merge vote counts and ratings with the movies dataset
movie_stats = qualified_movies.groupby('movieId').agg({'rating':'mean', 'userId':'count'}).reset_index()
movie_stats.columns = ['movieId', 'rating', 'vote_count']

# Merge the movie stats with the movies dataset to get titles and genres
movie_stats = movie_stats.merge(movies, on='movieId')

# Calculate the score using the weighted formula
movie_stats['score'] = movie_stats.apply(weighted_rating, axis=1)

# Sort the movies based on the score in descending order
top_movies = movie_stats.sort_values('score', ascending=False)

# Display the top 10 movies with title
print(top_movies[['title']].head(10))
```

```python
# Vectorize genres using TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['genres'])

# Calculate the cosine similarity between movies based on genres
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Create a function to get recommendations based on movie title
def get_content_recommendations(title, cosine_sim=cosine_sim):
    idx = movies[movies['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Get top 10 similar movies
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices]

# Example: Content-based recommendations for a given movie
get_content_recommendations('Godzilla (1998)')
```

```python
# Select top users and movies based on the number of ratings
top_users = ratings['userId'].value_counts().nlargest(100).index  # Top 1000 users
top_movies = ratings['movieId'].value_counts().nlargest(10000).index  # Top 10000 movies

# Filter ratings for these users and movies
filtered_data = ratings[ratings['userId'].isin(top_users)  & ratings['movieId'].isin(top_movies)]

# Merge filtered data with movie titles
filtered_data = filtered_data.merge(movies, on='movieId')

# Create pivot table with users as rows and movie titles as columns
user_movie_ratings = filtered_data.pivot_table(index='userId', columns='title', values='rating')

# Fill missing values with 0
user_movie_ratings.fillna(0, inplace=True)

# Train k-NN model
model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(user_movie_ratings)

# Function to get user recommendations
def get_user_recommendations(user_id, n_recommendations=10):
    if user_id not in user_movie_ratings.index:
        print(f"User {user_id} not found in the dataset.")
        return []
    
    # Reshape the user data for k-NN
    distances, indices = model_knn.kneighbors(user_movie_ratings.loc[user_id]
    .values.reshape(1, -1), n_neighbors=n_recommendations)
    
    # Extract recommended movie titles based on nearest neighbors
    recommended_movies = []
    for i in indices.flatten():
        recommended_movies.append(user_movie_ratings.columns[i])
    
    return recommended_movies

# Example: Get recommendations for a user
recommendations = get_user_recommendations(1)

# Convert to DataFrame and print
df = pd.DataFrame(recommendations, columns=['Movie Titles'])
print(df)
```

```python
# Transpose the matrix for item-based filtering
movie_user_ratings = user_movie_ratings.T

# Item-based collaborative filtering using k-NN
model_knn_movie = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn_movie.fit(movie_user_ratings)

def get_item_recommendations(movie_title, n_recommendations=10):
    distances, indices = model_knn_movie.kneighbors(movie_user_ratings.
    loc[movie_title].values.reshape(1, -1), n_neighbors=n_recommendations)
    recommended_movies = []
    for i in indices.flatten():
        recommended_movies.append(movie_user_ratings.index[i])
    return recommended_movies

# Example: Item-based recommendations for a given movie
exp=pd.DataFrame(get_item_recommendations('8 Seconds (1994)'),columns=['Movie Titles'])
print(exp)
```

```python
def hybrid_recommendation(user_id, movie_title, top_n=10):
    title_match = movies[movies['title'] == movie_title]
    
    if title_match.empty:
        print(f"Title '{movie_title}' not found in the dataset.")
        return []
    # User-based recommendations
    user_recs = set(get_user_recommendations(user_id, top_n))
    
    # Item-based recommendations
    item_recs = set(get_item_recommendations(movie_title, top_n))
    
    # Content-based recommendations
    content_recs = set(get_content_recommendations(movie_title))
    
    # Combine recommendations (intersection or union)
    final_recs = list(user_recs | item_recs | content_recs )[:top_n]
    
    return final_recs

# Example: Hybrid recommendations for a user and a movie
hybrid_recomendation=hybrid_recommendation(9, 'Action Jackson (1988)')
df1= pd.DataFrame(hybrid_recomendation, columns=['Movie Titles'])
print(df1)
```
